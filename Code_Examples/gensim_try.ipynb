```python
from gensim.test.utils import get_tmpfile
from gensim.models import Word2Vec
import pandas as pd

path = get_tmpfile("word2vec.model")
```

```python
filepath_train = "/home/mackenzie/workspace/PycharmProjects/DAADRISE_AbusiveLangProject/featureExtraction/EnglishCleanedTrainingData (1).csv"
train_data = pd.read_csv(filepath_train, dtype=str)
```

```python
sentences = train_data['cleaned_tweet']
model = Word2Vec([s.split() for s in sentences], size=100, window=5, min_count=5, workers=4)
```
