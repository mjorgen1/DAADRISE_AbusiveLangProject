```python
# plots for data
import matplotlib.pyplot as plt
import numpy as np
```

```python
eng_label = ['Hateful', 'Offensive', 'Neither']
germ_label = ['Abuse', 'Insult', 'Other']

eng_percents = [6, 77, 7]
germ_percents = [20.4, 11.9, 67.7]

# change as needed!
label = germ_label
values = germ_percents
```

```python
index = np.arange(len(label))

name_english = 'English Data-Set Class Distribution'
name_german = 'German Data-Set Class Distribution'

# change as needed!
plot_title = name_german
```

```python
def plot_bar_x(name):
    # this is for plotting purpose
    plt.bar(index, values)
    plt.xlabel('Labels', fontsize=13)
    plt.ylabel('Label % in Data-Set', fontsize=13)
    plt.xticks(index, label, fontsize=12, rotation=30) 
    plt.title(name, fontsize=20)
    fig_name = name + '.jpg'
    x1,x2,y1,y2 = plt.axis()
    plt.axis((x1,x2,0,100))
    for ind,data in enumerate(values):
        plt.text(x=ind , y =data+1 , s=f"{data}", fontsize=15)
    plt.tight_layout()
    plt.savefig(fig_name, dpi=600)
    plt.show()
```

```python
plot_bar_x(plot_title)
print('done!')
```
